## 梯度縮減範例

```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
#目標函數
def func(x): return np.square(x)
#目標函數一次微分
def dfunc(x): return 2 * x
```

#定義目標函數y=x*x

```
t=np.arange(-6.0,6.0,0.1)
print(t)
plt.plot(t,func(t), c='black')
#b預設藍色/black黑色
plt.show
```


# 定義梯度縮減函數
# 梯度：目標函數的一階導數(斜率)
# 梯度縮減：往斜率的"反方向"前進
# 在epochs次反覆運算中x的更新值¶
x_start: x的起始點
df: 目標函數的一階導函數
epochs: 反覆運算次數
lr: 學習率
回傳：x在每次反覆運算後的位置（包括起始點），長度為epochs+1

```
def GD(x_start,df, epochs, lr):
  xs=np.zeros(epochs+1)
  x=x_start
  xs[0]=x
  for i in range(epochs):
    dx=df(x) #v表示x的幅度
    v=-dx*lr #往斜率反方向前進，乘上learning rate
    x += v
    xs[i+1] =x
  return xs
```

```
# 起始權重
x_start = 5    
# 執行週期數
epochs = 15 
# 學習率   
lr = 0.8   

t = np.arange(-6.0, 6.0, 0.1)#起始數值(起始點,終點,間距)
print(t)

# 梯度下降法 
x = GD(x_start, dfunc, epochs, lr=lr)
print(x)
```

```
plt.plot(t, func(t), c='b')#畫出目標函數x*x#弧形藍色
plt.scatter(x, func(x), c='r')#畫出一階導數之值#點紅色
plt.plot(x, func(x), c='purple', label='lr={}'.format(lr))#畫出一階導數連線#線條紫色
plt. legend()
plt. show()
```

```
import matplotlib.pyplot as plt
```

#定義目標函數y=x*x

```
#目標函數
def func(x): return x*x
#目標函數一次微分
def dfunc(x): return 2*x
```

```
t=np.arange(-6.0,6.0,0.1)
print(t)
plt.plot(t,func(t), c='black')
plt.show
```

```
def GD(x_start,df, epochs, lr):
  xs=np.zeros(epochs+1)
  x=x_start
  xs[0]=x
  for i in range(epochs):
    dx=df(x)
    v=-dx*lr #往斜率反方向前進，乘上learning rate
    x += v
    xs[i+1] =x
  return xs
```

```
x_start=5
epochs=15
lr=0.8
x=GD(x_start, dfunc, epochs, lr)
print(x)
```

```
plt.plot(t, func(t), c='b')
plt.scatter(x, func(x), c='r')
plt.plot(x, func(x), c='r', label='lr={}'.format(lr))
plt. legend()
plt. show()
```
